# univirate visualization


```python
import matplotlib.pyplot as plt
```

```python
plt.hist(pred['age'], color='skyblue', edgecolor='black', bins=20)
plt.title('Age Distribution')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.hist(pred['bmi'], color='#091370', edgecolor='black', bins=20)
plt.title('BMI Distribution')
plt.xlabel('bmi')
plt.ylabel('frequency')
plt.show()
```

```python
plt.hist(pred['avg_glucose_level'], color='#1dad44', edgecolor='black', bins=15)
plt.title('Average Glucose Level Distribution')
plt.xlabel('Avg Glucose Level')
plt.ylabel('frequency')
plt.show()
```

```python
pred['gender'].value_counts().plot(kind='pie', autopct='%1.2f')
plt.plot()
```

# bivariate visualization

```python
# age vs avg_glucose_level scatter bar plot
plt.figure(figsize=(10,8))
plt.scatter(pred['age'], pred['avg_glucose_level'], color='#db541a', marker='+', alpha=0.6)
plt.grid()
plt.title("Age vs Averaeg Glucose Level")
plt.xlabel('age')    
plt.ylabel('avg_glucose_level')
plt.show
```

```python
#  age vs bmi scatter plot
plt.figure(figsize=(10,8))
plt.scatter(pred['age'], pred['bmi'], color='#8a0c45', marker='o', alpha=0.2)
plt.title('Age vs BMI')
plt.xlabel('age')
plt.ylabel('bmi')
plt.show()
```

```python
plt.hist(pred[pred['stroke']==0]['avg_glucose_level'], bins=20, alpha=0.7, label='No Stroke')
plt.hist(pred[pred['stroke']==1]['avg_glucose_level'], bins=20, alpha=0.7, label='Stroke')
plt.title('Avg Glucose Level by Stroke Status')
plt.xlabel('avg glucose level')
plt.ylabel('frquency')
plt.legend()
plt.show()
```
